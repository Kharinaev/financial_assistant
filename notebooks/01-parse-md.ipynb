```python
with open('/Users/alfa/Code/financial_assistant/data/raw/alfa_invest_begginer.md', 'r') as f:
    alfa_invest_begginer_text = f.readlines()

with open('/Users/alfa/Code/financial_assistant/data/raw/alfa_invest_advanced.md', 'r') as f:
    alfa_invest_advanced_text = f.readlines()
```

```python
for i in range(len(alfa_invest_advanced_text)):
    alfa_invest_advanced_text[i] = alfa_invest_advanced_text[i].replace('\xa0', ' ')

for i in range(len(alfa_invest_begginer_text)):
    alfa_invest_begginer_text[i] = alfa_invest_begginer_text[i].replace('\xa0', ' ')
```

```python
alfa_invest_advanced_text[0]
```

```python
advanced_articles = []

cur_lines = []
cur_article_name = None
for line in alfa_invest_advanced_text:
    if "~" in line and "**" in line:
        if cur_article_name is not None:
            advanced_articles.append({
                "heading" : cur_article_name,
                "decription" : "",
                "text" : "\n".join(cur_lines) 
            })
        cur_article_name = line.replace('#','').replace('*','').replace('~','').strip()
        cur_lines = []
    elif len(line.strip()) > 0:
        cur_lines.append(line.strip())
```

```python
begginer_articles = []

cur_lines = []
cur_article_name = None
status = None # desc or content
cur_description = None
for line in alfa_invest_begginer_text:
    # print(line.replace(' ','')[:4])
    if "~" in line and "#" in line:
        if cur_article_name is not None:
            begginer_articles.append({
                "heading" : cur_article_name,
                "decription" : cur_description,
                "text" : "\n".join(cur_lines) 
            })
        cur_article_name = line.replace('#','').replace('*','').replace('~','').strip()
        cur_lines = []
    elif "[description]" in line:
        status = 'desc'
    elif "[content]" in line:
        status = 'content'
    elif status == 'desc' and len(line.strip()) > 0:
        cur_description = line.strip()
    elif len(line.strip()) > 0:
        cur_lines.append(line.strip())
```

# Добавляем идентификаторы статей

```python
for idx, article in enumerate(begginer_articles):
    article["article_id"] = f"AIB_{idx:04d}"
```

```python
for idx, article in enumerate(advanced_articles):
    article["article_id"] = f"AIA_{idx:04d}"
```

# Бьем на чанки по заголовкам

```python
begginer_chunks = []
for idx, article in enumerate(begginer_articles):
    lines = article['text'].split('\n')
    cur_chunk_lines = []
    cur_chunk_header = article["heading"]
    for line in lines:
        if line.strip()[0]=='#':
            if len(cur_chunk_lines) > 0:
                begginer_chunks.append({
                    "article_id" : article["article_id"],
                    "chunk_id" : f"{article['article_id']}_{len(begginer_chunks)+1:04d}",
                    "heading": cur_chunk_header,
                    "text": "\n".join(cur_chunk_lines)
                })
            cur_chunk_header = line.strip()
            cur_chunk_lines = []
        else:
            cur_chunk_lines.append(line)
```

```python
advanced_chunks = []
for idx, article in enumerate(advanced_articles):
    lines = article['text'].split('\n')
    cur_chunk_lines = []
    cur_chunk_header = article["heading"]
    for line in lines:
        if line.strip()[0]=='#':
            if len(cur_chunk_lines) > 0:
                advanced_chunks.append({
                    "article_id" : article["article_id"],
                    "chunk_id" : f"{article['article_id']}_{len(advanced_chunks)+1:04d}",
                    "heading": cur_chunk_header,
                    "text": "\n".join(cur_chunk_lines)
                })
            cur_chunk_header = line.strip()
            cur_chunk_lines = []
        else:
            cur_chunk_lines.append(line)
```

```python
len(begginer_chunks), len(advanced_chunks)
```

```python
lengths_beg = [len(c['text'].split()) for c in begginer_chunks]
lengths_adv = [len(c['text'].split()) for c in advanced_chunks]
```

# Сохраняем

```python
beg_articles_df = pd.DataFrame(begginer_articles).set_index('article_id')
adv_articles_df = pd.DataFrame(advanced_articles).set_index('article_id')

beg_chunks_df = pd.DataFrame(begginer_chunks).set_index('chunk_id')
adv_chunks_df = pd.DataFrame(advanced_chunks).set_index('chunk_id')
```
